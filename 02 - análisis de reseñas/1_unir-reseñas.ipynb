```python
import pandas as pd
import openpyxl
```

```python
google_reviews = pd.read_excel('google.xlsx')
tripadvisor_reviews = pd.read_excel('tripadvisor.xlsx')
```

```python
print(google_reviews.info())
print(tripadvisor_reviews.info())
```

```python

df_google = google_reviews[['review_rating', 'review_text', 'review_timestamp']]
df_google["date"] = pd.to_datetime(df_google["review_timestamp"], unit="s")
df_google["date_str"] = df_google["date"].dt.strftime("%Y-%m-%d")
df_google.rename(columns={"review_rating": "rating", "review_text": "comentario", "date_str": "fecha"}, inplace=True)
df_tripadvisor = tripadvisor_reviews[['review_rating', 'review_text', 'review_date']]
df_tripadvisor.rename(columns={"review_rating": "rating", "review_text": "comentario", "review_date": "fecha"}, inplace=True)

df_tripadvisor["fuente"] = "tripadvisor"
df_google["fuente"] = "google_maps"
```

```python
df_total = pd.concat([df_google, df_tripadvisor], ignore_index=True)
print(df_total.head())
```

```python
df_total.to_csv("reseñas_unificadas.csv", index=False, encoding="utf-8")
print("Total reseñas:", len(df_total))
```
